```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
```

```python
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_Final','Var. Dia (%)':'Var_Dia_Pct'}).copy()
df_principal
```

```python
##Criando colunas calculadas com data frame
df_principal['Var_pct'] = df_principal['Var_Dia_Pct'] / 100
df_principal['Valor_Inicial'] = df_principal['Valor_Final'] / (df_principal['Var_pct'] +1)
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_Dia_Pct'] / 100
df_principal
```

```python
##merge columns de quantidade teorica
df_principal = df_principal.merge(df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
#excluir a coluna código que veio do merge
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
#Renomear a coluna
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_Teorica'}).copy()
df_principal
```

```python
##Criar a coluna Variação %
df_principal['Variacao_Reais'] = (df_principal['Valor_Final'] - df_principal['Valor_Inicial']) * df_principal['Qtd_Teorica']
df_principal
```

```python
df_principal
```

```python
#formatar os tipos de float, tirando a notação cientifica
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
#formatar datatype da coluna para inteiro
df_principal['Qtd_Teorica'] = df_principal['Qtd_Teorica'].astype(float).astype(int)
df_principal
#print(df_principal.dtypes)
```

```python
#criar coluna calculada variação usando o chatgpt
#df_principal['Resultado'] = df_principal['Variacao_Reais'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal['Resultado'] = pd.Series(['Subiu' if x > 0 else 'Desceu' if x < 0 else 'Estavel' for x in df_principal['Variacao_Reais']])

df_principal
```

```python
valor = "Desceu"

resultado = df_principal[df_principal['Resultado'] == valor]

print(resultado)
```

```python
#criar outro merge para trazer o nome da empresa - aba ticker e dps excluir a coluna q ira se repetir
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
#criar outro merge para trazer o seguimento - aba chatgpt e dps excluir a coluna q ira se repetir
df_principal = df_principal.merge(df_ChatGPT, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
#Renomear a coluna
df_principal = df_principal.rename(columns={'Idade (em anos)':'Idade'}).copy()
df_principal
```

```python
#Criar a coluna de categoria de idade usando o chatgpt
#df_principal['Cat_Idade'] = pd.Series(['Mais de 100 anos' if x > 100 else 'Menos de 50' if x < 50 else 'Entre 50 e 100 anos' for x in df_principal['Idade']])
df_principal['Cat_Idade'] = df_principal['Variacao_Reais'].apply(lambda x: 'Mais de 100 anos' if x > 0 else ('Menos de 50' if x < 0 else 'Entre 50 e 100 anos'))
df_principal
```

```python
#checando se existem registros com o valor subiu na coluna resultados
valor = "Subiu"

resultado = df_principal[df_principal['Resultado'] == valor]

print(resultado)
df_principal['Resultado']
```

```python
##Criando os calculos da aba analises

# Calculando o maior valor
maior = df_principal['Variacao_Reais'].max()

# Calculando o menor valor
menor = df_principal['Variacao_Reais'].min()

# Calculando a média
media = df_principal['Variacao_Reais'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_Reais'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == "Desceu"]['Variacao_Reais'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
##Criando um novo dataframe so com os valores que subiram

df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
#criando outro dataframe agrupando as somas dos dque subiram por seguimento. É necessário resetar o index
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_Reais'].sum().reset_index()
df_analise_segmento
```

```python
#criando outro dataframe agrupando a soma de toda a coluna resultado por saldo. É necessário resetar o index
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_Reais'].sum().reset_index()
df_analise_saldo
```

```python
#utilizando o plotly.express para fazer os graficos
#criando um grafico de barras

fig = px.bar(df_analise_saldo, x='Resultado',
             y='Variacao_Reais',
             #text=['{:.2}B'.format(x / 1000000000) for x in df_analise_saldo['Variacao_Reais'][:10]],
             text_auto='.3~s',
             title='Variacao Reais por Resultado')

#para imprimir o grafico
fig.show()
```

```python
fig.update_layout(yaxis_tickprefix = '$', yaxis_tickformat = ',.2f')

fig.show()
```
